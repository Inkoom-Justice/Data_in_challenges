# Challenge 6: Investor Funds

### Scenario
You are a financial data analyst on Wall Street and you have been tasked with summarizing data about investor funds. Your manager would like the following questions answered.

```python
# import libraries/packages
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# get data
url = 'https://raw.githubusercontent.com/datasets/investor-flow-of-funds-us/master/data/weekly.csv' 
df = pd.read_csv(url)
df.head()
```

```python
df.info()
```

### 1. What is the total equity investment over time

```python
# Plotting the total equity investment over time
plt.figure(figsize=(12, 6))
plt.plot(df['Date'], df['Total Equity'])
plt.xlabel('Date')
plt.ylabel('Total Equity Investment')
plt.title('Total Equity Investment Over Time')
plt.xticks(rotation = 60)
plt.show()
```

### 2. How are the different types of funds (equity, bond, hybrid) distributed

```python
import matplotlib.pyplot as plt

labels = ['Equity', 'Bond', 'Hybrid']
sizes = [df['Total Equity'].sum(), df['Total Bond'].sum(), df['Hybrid'].sum()]

plt.figure(figsize=(8, 6))
plt.bar(labels, sizes, color=['#ff9999', '#66b3ff', '#99ff99'])
plt.xlabel('Fund Type')
plt.ylabel('Total Value')
plt.title('Fund Distribution')

for index, value in enumerate(sizes):
    plt.text(index, value, f'{value:.2f}')

plt.show()
```

### 3. What is the trend of domestic and world equity investments?

```python
# Plotting the trend of domestic and world equity investments
plt.figure(figsize=(12, 6))
plt.plot(df['Date'], df['Domestic Equity'], label='Domestic Equity')
plt.plot(df['Date'], df['World Equity'], label='World Equity')
plt.xlabel('Date')
plt.ylabel('Equity Investment')
plt.title('Trend of Domestic and World Equity Investments')
plt.legend()
plt.xticks(rotation=60)
plt.show()
```

### 4. How do bond investments compare between taxable and municipal bonds?

```python
# Creating a bar chart to compare taxable and municipal bond investments
labels = ['Taxable Bond', 'Municipal Bond']
heights = [df['Taxable Bond'].sum(), df['Municipal Bond'].sum()]

plt.figure(figsize=(8, 6))
plt.bar(labels, heights)
plt.xlabel('Bond Type')
plt.ylabel('Bond Investment')
plt.title('Comparison of Taxable and Municipal Bond Investments')
plt.show()
```

You are a data analyst at one of the worldâ€™s largest wine companies and you have been tasked with summarizing data about wine. Your manager would like the following questions answered.

```python
# import libraries/packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# get dataset
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data' 
wine = pd.read_csv(url, header=None)
wine.head()
```

### 1. Delete the first, fourth, seventh, nineth, eleventh, thirteenth and fourteenth columns.

```python
# delete the first, fourth, seventh, nineth, eleventh, thirteenth and fourteenth columns
wine_1 = wine.drop([0,3,6,8,10,12,13], axis=1)
wine_1.head()
```

### 2. Assign the columns as below: alcohol, malic_acid, alcalinity_of_ash, magnesium, flavanoids, proanthocyanins, hue.

```python
# create an array of column names
column_headers = ['alcohol', 'malic_acid', 'alcalinity_of_ash', 'magnesium', 'flavanoids', 'proanthocyanins', 'hue']

# assign column names
wine_1.columns = column_headers
wine_1.head()
```

### 3. Set the values of the first 3 rows in the alcohol column as NaN.

```python
# set values of first 3 rows in alcohol column as NaN
wine_1.loc[:2, 'alcohol'] = np.nan
wine_1.head()
```

### 4. Now set the value of the rows 3 and 4 of the magnesium column as NaN

```python
# set rows 3 and 4 of magnesium column as NaN
wine_1.loc[2:3, 'magnesium'] = np.nan
wine_1.head()
```

### 5. Fill in the null values (NaN) with the number 10 in the alcohol column and 100 in magnesium column.

```python
# fill in null values
wine_1.loc[:2, 'alcohol'] = 10
wine_1.loc[2:3, 'magnesium'] = 100
wine_1.head()
```

### 6. Count the number of missing values in the entire dataset.

```python
# missing values count
missing_values_count = wine_1.isna().sum().sum()
print('The number of missing values in the entire dataset is', missing_values_count)
```
